In this particular binary classification case, the log odds is given by:

$$logit(\sigma) \ \ = \ \ w^Tx \ \ = \ \ ln \dfrac{\sigma}{1 - \sigma} \ \ = \ \ ln \dfrac{p(y=1|x,w)}{p(y=0|x,w)}$$

Thus the logistic function for $w^Tx$ is given by:

$$logistic(w^Tx) \ \ = \ \ \sigma (w^Tx) \ \ = \ \ logit^{-1}(w^Tx) \ \ = \ \ \dfrac{1}{1 + exp(-w^Tx)} \ \ = \ \ \dfrac{exp(w^Tx)}{exp(w^Tx) + 1}$$

Let the probability that the canine is infected be $p(y=1|x,w)$ and the probability that the canine is not infected be $p(y=0|x,w)$. Probability values must range from 0 to 1 and the sum of all possible individual probabilities of a set of events must add up to 1. Therefore, the probability that the canine is not infected is $1 - p(y=1|x,w)$.

Probability that canine is infected:

$$p(y=1|x,w) \ \ = \ \ \hat{y} \ \ = \ \ \dfrac{exp(w^Tx)}{exp(w^Tx) + 1}$$

Probability that canine is not infected:

$$p(y=0|x,w) \ \ = \ \ 1 - \hat{y} \ \ = \ \ 1 - \dfrac{exp(w^Tx)}{exp(w^Tx) + 1} \ \ = \ \ \dfrac{1}{exp(w^Tx) + 1}$$

The objective function for classification is Cross Entropy (CE). Typically CE is calculated as follows:

$$L_{CE} \ \ = \ \ L(w) \ \ = \ \ - \dfrac{1}{M} \sum\limits_{i=1}^{m} \left[y_i \log(\hat{y}_i) + (1-y_i) \log(1 - \hat{y}_i)\right]$$

For Logistic Regression, I will assume specifically, for two classes (1 that the canine is infected and 0 is not infected with the parasite), Binary Cross Entropy (BCE). Therefore the loss function of $w$ is as follows:

$$L_{BCE} \ \ = \ \ L(w) \ \ = \ \ - \sum\limits_{i=1}^{m} \left[y_i \log(\hat{y}_i) + (1-y_i) \log(1 - \hat{y}_i)\right]$$


Similarly, it can be written out as:

$$= \ \ - \sum\limits_{i=1}^{m} \left[y_i \log(p(y=1|x,w)) + (1-y_i) \log(p(y=0|x,w))\right]$$

Or:

$$= \ \ - \sum\limits_{i=1}^{m} \left[y_i \log(\dfrac{exp(w^Tx)}{exp(w^Tx) + 1}) + (1-y_i) \log(\dfrac{1}{exp(w^Tx) + 1})\right]$$

The BCE loss function measures the discrepancies between the predicted $\hat{y}_i$ and the ground truth (actual) $y_i$ binary labels and heavily penalizes confident wrong predictions by assigning a high loss value when $\hat{y}_i$ label is far from the $y_i$ label. Being the case, a loss approaching zero can indicate that the $\hat{y}_i$ label is closer to the $y_i$ label.

As stated in Question 3, for Logistic Regression, I will assume specifically Binary Cross Entropy (BCE) and the loss function of $w$ is as follows:

$$L_{BCE} \ \ = \ \ L(w) \ \ = \ \ - \sum\limits_{i=1}^{m} \left[y_i \log(\hat{y}_i) + (1-y_i) \log(1 - \hat{y}_i)\right]$$

It is also known that $\hat{y} \ \ = \ \ \dfrac{exp(w^Tx)}{exp(w^Tx) + 1} \ \ = \ \ \dfrac{1}{1 + exp(-w^Tx)}$

Therefore the derivation of the BCE loss function, $\nabla_w L_{BCE}$, is calculated as follows:

$$\frac{\partial L(w)}{\partial w} \ \ = \ \ \frac{\partial L(w)}{\partial \hat{y}} \cdot \frac{\partial \hat{y}}{\partial w^Tx} \cdot \frac{\partial w^Tx}{\partial w}$$

---

Step 1: deriving $\frac{\partial L(w)}{\partial \hat{y}}$

$$\frac{\partial L(w)}{\partial \hat{y}} \ \ = \ \ -(\dfrac{y}{\hat{y}} - \dfrac{1 - y}{1 - \hat{y}})$$

$$= \ \ \dfrac{\hat{y} - y}{\hat{y}(1-\hat{y})}$$

---

Step 2: deriving $\frac{\partial \hat{y}}{\partial w^Tx}$

$$\frac{\partial \hat{y}}{\partial w^Tx} \ \ = \ \ \frac{\partial}{\partial w^Tx} \hat{y}_i \ \ = \ \ \frac{\partial}{\partial w^Tx} \left[\dfrac{1}{1+exp(-w^Tx)}\right] \ \ = \ \ \frac{\partial}{\partial w^Tx} (1 + exp(-w^Tx))^{-1}$$

$$= \ \ -(1 + exp(-w^Tx))^{-2}(-exp(-w^Tx)) \ \ = \ \ \dfrac{exp(-w^Tx)}{1 + exp(-w^Tx)^2}$$

$$= \ \ \dfrac{1}{1+exp(-w^Tx)} \left(\dfrac{1+exp(-w^Tx)}{1+exp(-w^Tx)} - \dfrac{1}{1+exp(-w^Tx)}\right)$$

$$= \ \ \dfrac{1}{1+exp(-w^Tx)} \left(1 - \dfrac{1}{1+exp(-w^Tx)}\right)$$

$$= \ \ \hat{y}(1-{\hat{y}})$$

---

Step 3: deriving $\frac{\partial w^Tx}{\partial w}$

$$\frac{\partial w^Tx}{\partial w} \ \ = \ \ x$$

---

Step 4:

$$\frac{\partial L(w)}{\partial w} \ \ = \ \ \dfrac{\hat{y} - y}{\hat{y}(1-\hat{y})} \cdot \hat{y}(1-{\hat{y}}) \cdot x \ \ = \ \ (\hat{y} - y)x$$

Thus the derivative (gradient) minimizing the loss function with respect to $w$ is:

$$\nabla_w L_{BCE} \ \ = \ \ \sum\limits_{i=1}^{m} (\hat{y_i} - y_i)x_i$$

### Question 5 - Imbalanced dataset (10 points)

You are now told that in the dataset  

$$p(y=0) >> p(y=1)$$

Can you comment if the accuracy of Logistic Regression will be affected by such imbalance?



Answer:

The case that $p(y=0) >> p(y=1)$ signifies that the $y=0$ (where the canine is not infected) is the majority class greatly outweighing the minority class $y=1$ (where the canine is infected). This class imbalance affects Logistic Regression in that it tends to be biased towards the majority class $y=0$, thus affecting the accuracy of predicting the $y=0$ correctly over $y=1$. In other words, due to the class imbalance the logistic regressor would do poorly in capturing the patterns and characteristics of the minority class, however, still a model could have high accuracy scores due to prioritizing prediction of $y=0$ for almost all instances over $y=1$.

Therefore, accuracy score alone will not be a good indicator of model performance as it can be misleading due to the class imbalance, therefore, metrics such as precision, recall, and F1 score (harmonic mean of precision and recall) should be considered as well.

```python
#import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix

# documentation on SGDClassifier can be found here
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.SGDClassifier.html
# setting loss='log_loss' gives logistic regression
from sklearn.linear_model import SGDClassifier
```

```python
# load and preview the dataset
df = pd.read_csv('CBC_data.csv')
df.head()
```

```python
# view data info
df.info()
```

```python
# view count of classes
df['PARASITE_STATUS'].value_counts()
```

It can be seen that there is class imbalance - the counts of class "Negative" greatly outweights counts of class "Positive". I had stated my answer in Question 5 regarding class imbalances. Therefore, I will guess that, in this case for the canine dataset, my accuracy score will be high and my F1 score will be low.

```python
# determine if there are NULL values
df.isna().sum()
```

```python
# replace NULLs with averages
df = df.fillna(df.mean(numeric_only=True))
df.isna().sum()
```

```python
# transform the features from categorical to numerical
transform = LabelEncoder()
df["ID"] = transform.fit_transform(df['ID'])
df["SEX"] = transform.fit_transform(df['SEX'])
df['TYPEAREA'] = transform.fit_transform(df['TYPEAREA'])
df['SEX.REPRO'] = transform.fit_transform(df['SEX.REPRO'])
df['REPRO.STATUS'] = transform.fit_transform(df['REPRO.STATUS'])
df['PARASITE_STATUS'] = transform.fit_transform(df['PARASITE_STATUS'])
```

```python
# set up X, y
X = df.drop('PARASITE_STATUS', axis=1)
y = df['PARASITE_STATUS']
```

```python
random_state = 123

# split dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)
```

```python
# scale all features to have mean of 0 and standard deviation of 1
sc = StandardScaler()
X_train = pd.DataFrame(sc.fit_transform(X_train), index=X_train.index, columns=X_train.columns)
X_test = pd.DataFrame(sc.transform(X_test), index=X_test.index, columns=X_test.columns)

X_train.describe().round(5)
```

```python
# set hyperparameters and cross validation
learning_rate = 0.01
max_iter = 100
cv = 5

# use log_loss for logistic regression and l2 regularization
model = SGDClassifier(loss='log_loss', penalty='l2', learning_rate='constant', eta0=learning_rate, max_iter=max_iter, random_state=random_state)

# perform cross-validation and print the F1 scores
f1_scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='f1_macro')
print(f"Cross-Validation F1 Scores:\n {f1_scores}")
print()

# train the model on the full training set with selected hyperparameters
model.fit(X_train, y_train)

# evaluate performance on test set
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
print(f"\nTest Set Accuracy Score:\n {acc}")
print(f"\nTest Set F1 Score:\n {f1}")
```

```python
# print Confusion Matrix and Classification Report
cmatrix = confusion_matrix(y_test, y_pred)
clfreport = classification_report(y_test, y_pred)
print(f"Confusion Matrix:\n {cmatrix}")
print(f"\nClassification Report:\n {clfreport}")
```

As guessed, my model's accuracy score is higher than the F1 score.

```python
# plot F1 score vs. iteration number
print(f"F1 Score and Accuracy at each iteration:")

# initialize accuracy and f1 scores
accuracy_scores_iter = []
f1_scores_iter = []

# print scores at each iteration
for t in range(0, max_iter + 1):
    model.partial_fit(X_train, y_train)#, classes=np.unique(y_train))
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    accuracy_scores_iter.append(acc)
    f1_scores_iter.append(f1)

    print(f"Iteration {t}: F1 Score = {f1}, Accuracy: {acc}")
    print()

print()
print(f"Plot the F1 Score vs Iteration Number t")

plt.figure(figsize=(10, 8))

plt.plot(range(0, max_iter + 1), f1_scores_iter)
plt.xlabel('Iteration Number (t)')
plt.ylabel('F1 Score')
plt.title('F1 Score vs. Iteration Number')
plt.show()
```

I wanted to view both F1 score and accuracy score on one plot.

```python
# plot accuracy and f1 scores
plt.figure(figsize=(10, 8))

plt.plot(range(0, max_iter + 1), f1_scores_iter, label='F1 Score')
plt.plot(range(0, max_iter + 1), accuracy_scores_iter, label='Accuracy Score')
plt.xlabel('Iteration Number (t)')
plt.ylabel('Scores')
plt.title('Scores (Accuracy and F1) vs. Iteration Number')
plt.legend()
plt.show()
```

As guessed, I have high accuracy scores of 0.927 but low F1 scores of 0.043. Due to the class imbalance the model is prioritizing predicting the Negative class (the majority class) correctly over the Positive class (minority class) - as shown in the high accuracy scores. As stated in Question 3, the loss function heavily penalizes confident wrong decisions, thus showing low F1 scores - indicating poor performance by the model to correctly identify the Positive (minority) class in this binary classification task.
