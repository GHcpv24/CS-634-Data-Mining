## Question 2 (25 points)

A sequential experiment involves repeatedly drawing a ball from one of the two urns, noting the number on the ball and replacing the ball in the urn. Urn 0 contains a ball with the number 0 and two balls with the number 1. Urn 1 contains five balls with the number 0 and one ball with the number 1. 

The urn from which the first ball is drawn is selected by flipping a fair coin. Urn 0 is used if the outcome is H and urn 1 is used if the outcome is T. **The urn used in a subsequent draws corresponds to the number on the ball drawn in the previous draw.** 

What is the probability of a specific sequence of the numbers on drawn balls being 0011 ? 

Type the answer here using the [latex syntax](https://wch.github.io/latexsheet/) or handwrite the answer, upload the picture in the same folder and use a new markdown cell with markdown syntax `![title](image_name.png)`


Let's first start with what is known:

* Urn 0 contains one ball with number 0 and two balls with number 1. Then the probability of drawing a ball with number 0 for Urn 0 is $\dfrac{1}{3}$, and the probability of drawing a ball with number 1 for Urn 0 is $\dfrac{2}{3}$.

* Urn 1 contains five balls with number 0 and one ball with number 1. Then the probability of drawing a ball with number 0 for Urn 1 is $\dfrac{5}{6}$, and the probability of drawing a ball with number 1 for Urn 1 is $\dfrac{1}{6}$.

* The probability that the first draw to select Urn 0 from the coin toss (outcome = Heads) is $\dfrac{1}{2}$, and the probability of the first draw for Urn 1 (Tails) is also $\dfrac{1}{2}$

* The Urn used in the subsequent draws correspond to the number on the ball from the previous draw.

The probability of a specific sequence of numbers on the balls drawn being 0011 assumes that the first ball drawn was numbered 0 (could be from either Urn 0 or Urn 1), then Urn 0 is used for the second draw with the second ball having a number 0, then Urn 0 is used for the third draw with the third ball having a number of 1, and lastly Urn 0 is used for the fourth draw with the fourth ball having a number 1.

Let Probability = Pr, H = selecting Urn 0, A = Urn 0, B = Urn 1, a = ball with number 0, and b = ball with number 1.

If the outcome of the coin toss is Heads, the first Urn to be drawn from is Urn 0. In this case, the probability of a specific sequence of numbers on the balls drawn being 0011 starting with Urn 0 is calculated as follows:

Pr(0011|H) = Pr(H) * Pr(a|A) * Pr(a|A) * Pr(b|A) * Pr(b|B)

= $\dfrac{1}{2} * \dfrac{1}{3} * \dfrac{1}{3} * \dfrac{2}{3} * \dfrac{1}{6}$

= $\dfrac{2}{324}$

$\approx$ 0.00617 probability if Urn 0 is selected for the first draw

If the outcome of the coin toss is Tails, the first Urn to be drawn from is Urn 1. In this case, the probability of a specific sequence of numbers on the balls drawn being 0011 starting with Urn 1 is calculated as follows:

Pr(0011|T) = Pr(T) * Pr(a|B) * Pr(a|A) * Pr(b|A) * Pr(b|B)

= $\dfrac{1}{2} * \dfrac{5}{6} * \dfrac{1}{3} * \dfrac{2}{3} * \dfrac{1}{6}$

= $\dfrac{10}{648}$

$\approx$ 0.0154 probability if Urn 1 is selected for the first draw

Thus:

Probability(0011) = Pr(0011|H) + Pr(0011|T)

= 0.00617 + 0.0154 = 0.02157 probability

There is a 0.02157 (or 2.157%) probability that for a specific sequence of numbers on the balls drawn to be 0011.

```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set mean and covariance for bivariate Gaussian distribution
mean = np.array([0,-1])
cov = np.array([[1,-0.5],[-0.5,1]])

np.random.seed(123)

# Cholesky factorization
Chol = np.linalg.cholesky(cov)

samples = np.random.normal(size=(2500,2))
x, y = (np.dot(samples, Chol.T) + mean).T

plt.figure(figsize=(10, 8))
plt.scatter(x, y)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Bivariate Gaussian Distribution')
plt.show()
```

```python
# lambda (mean) value for Poisson distribution
lam = np.mean(df['Brooklyn Bridge'].values)

# simulated counts
sim = np.random.poisson(lam, len(df['Brooklyn Bridge'].values))

plt.figure(figsize=(10, 8))

plt.hist(sim, edgecolor='black')
plt.xlabel('Number of Cyclists')
plt.ylabel('Frequency')
plt.title('Simulated Number of Cyclists that Cross the Brooklyn Bridge Per Day')
plt.show()
```
